Univariate Analysis

```python
sns.set_style("whitegrid")
num_cols = ['visits', 'age', 'income', 'children', 'school', 'health1', 'health2', 'access']

fig, axes = plt.subplots(len(num_cols), 1, figsize=(10, 20))

for i, col in enumerate(num_cols):
  sns.histplot(Data[col], bins=30, kde=True, ax=axes[i])
  axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
num_cols = ['visits', 'age', 'income', 'children', 'school', 'health1', 'health2', 'access']

fig, axes = plt.subplots(len(num_cols), 1, figsize=(10, 20))

for i, col in enumerate(num_cols):
  sns.boxplot(x=Data[col], ax=axes[i], color="skyblue")
  axes[i].set_title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(len(num_cols), 1, figsize=(10, 20))

for i, col in enumerate(num_cols):
  sns.violinplot(x=Data[col], ax=axes[i], color="lightcoral")
  axes[i].set_title(f'Violin Plot of {col}')

plt.tight_layout()
plt.show()
```

Bivariate Analysis

```python
sns.set_style("whitegrid")
num_cols = ['visits', 'age', 'income', 'children', 'school', 'health1', 'health2', 'access']
cat_cols = ['married', 'gender', 'ethnicity', 'enroll', 'program']

fig, axes = plt.subplots(3, 2, figsize=(12, 12))
pairs = [('age', 'income'), ('visits', 'income'), ('children', 'income'), ('school', 'income'), ('health1', 'health2')]

for i, (x, y) in enumerate(pairs):
  sns.scatterplot(x=Data[x], y=Data[y], ax=axes[i//2, i%2], alpha=0.6)
  axes[i//2, i%2].set_title(f'{x} vs {y}')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 18))

for i, cat in enumerate(cat_cols):
    if i < 5:
        sns.boxplot(x=Data[cat], y=Data['visits'], ax=axes[i//2, i%2])
        axes[i//2, i%2].set_title(f'Visits by {cat}')
        axes[i//2, i%2].set_xticklabels(axes[i//2, i%2].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(Data[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

Multivariate Analysis

```python
sns.set_style("whitegrid")

num_cols = ['visits', 'age', 'income', 'children', 'school', 'health1', 'health2', 'access']
cat_cols = ['married', 'gender', 'ethnicity', 'enroll', 'program']


sns.pairplot(Data, vars=['visits', 'age'], hue='gender', diag_kind='hist', palette='coolwarm')
plt.show()
```

```python
# Heatmap Showing Interactions Between Categorical and Numerical Variables
df_encoded = Data.copy()
df_encoded[cat_cols] = Data[cat_cols].apply(lambda x: pd.factorize(x)[0])  # Convert categorical to numerical

plt.figure(figsize=(10, 6))
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Interaction Heatmap of Categorical & Numerical Variables")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='enroll', y='visits', hue='gender', data=Data, split=True, palette="muted")
plt.title("Visits Distribution by Enrollment Status and Gender")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Data['age'], Data['income'], Data['visits'], c= Data['visits'], cmap='coolwarm', alpha=0.7)
ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_zlabel('Visits')
ax.set_title("3D Scatter Plot of Age, Income, and Visits")
plt.show()
```
